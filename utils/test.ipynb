1. 对爬取的牵牛花图像做图像增强：

```python
import img_enhance 
import os
import cv2
```

```python
src = 'D:/workspace/down_img/from_teacher/plumeria_ubra'
dst = 'E:/flowers_dataset/plumeria_ubra'
for i in os.listdir(src):
    name = src + '/' + i 
    img = cv2.imread(name)
    img2 = img_enhance.imgCrop(img)
    cv2.imwrite(dst + '/' + i, img2)
    print(i + '裁剪完成')
```
